```python
# Preload all NRH files by frequency
def load_nrh_by_freq(nrh_paths):
    out = {}
    for p in nrh_paths:
        parts = p.split('_')
        try:
            freq = int(parts[-2])  # 150 MHz etc.
        except ValueError:
            continue

        out[freq] = readsav(p, python_dict=True)
    return out

nrh_by_freq = load_nrh_by_freq(nrh_paths)
frequencies = sorted(nrh_by_freq.keys())

# A helper function to build the SunPy Map for one image
def build_map(nrh_file, index):
    hdr = nrh_file['nrh_hdr']
    data = nrh_file['nrh_data'][index]

    b = {name: hdr[name][index] for name in hdr.dtype.names}

    ref = SkyCoord(
        0*u.arcsec, 0*u.arcsec,
        frame=frames.Helioprojective(observer='earth', obstime=b['DATE_OBS'])
    )

    header = sunpy.map.make_fitswcs_header(
        data,
        ref,
        reference_pixel=[int(b['CRPIX1'])-1, int(b['CRPIX2'])-1]*u.pixel,
        scale=[float(b['CDELT1']), float(b['CDELT2'])]*u.arcsec/u.pixel,
        wavelength=float(b['FREQ'])*u.MHz
    )

    return sunpy.map.Map(data, header)

# Plot all frequencies for a single timestamp
import matplotlib.cm as cm
import matplotlib.patches as mpatches

cmap = cm.get_cmap('rainbow', len(frequencies))

def plot_timestamp(index, contour_level, save_path=None):
    fig = plt.figure(figsize=[7,7])

    # Build the map from the FIRST frequency to get the WCS
    f0 = frequencies[0]
    base_map = build_map(nrh_by_freq[f0], index)
    ax = fig.add_subplot(111, projection=base_map)

    patches = []

    for i, f in enumerate(frequencies):
        nrh_file = nrh_by_freq[f]
        m = build_map(nrh_file, index)
        data = m.data
        data_max = np.nanmax(data)
        thr = data_max * (contour_level / 100)
        levels = [thr, data_max]

        with frames.Helioprojective.assume_spherical_screen(m.observer_coordinate):
            cf = ax.contourf(
                data,
                levels=levels,
                cmap=mcolors.ListedColormap(cmap(i)),
                alpha=0.5
            )

        # Extract actual color
        col = cf.collections[0].get_facecolor()[0]
        patches.append(mpatches.Patch(color=col, label=f'{f} MHz'))

    ax.set_title(m.meta['date-obs'])
    ax.set_xlabel('Helioprojective Longitude (arcsec)')
    ax.set_ylabel('Helioprojective Latitude (arcsec)')
    ax.legend(handles=patches, loc='upper right')
    base_map.draw_limb(color='k')
    ax.grid(False)
    fig.tight_layout()

    if save_path:
        fig.savefig(f'{save_path}/nrh_{index:04d}.png', dpi=300)
        plt.close(fig)
    else:
        plt.show()


# Export all timestamps
n_timestamps = 2400
out_dir = f'{data_dir}/png/nrh/20251006'

import os
os.makedirs(out_dir, exist_ok=True)

for ts in range(n_timestamps):
    print(f'Plotting {ts}/{n_timestamps}')
    plot_timestamp(ts, contour_level=80, save_path=out_dir)
```

```python
def get_nrh_file(paths, f_mhz):
    """
    Find the full path of the NRH file 
    with a given frequency in a list of paths.
    """
    for p in paths:
        parts = p.split('_')
        try:
            freq = int(parts[-2])
        except ValueError:
            continue
        if freq == f_mhz:
            return p
    return None
```

```python
# Est. major axis of the beam size @ NRH freqs
def calc_major_axis(x1=0, y1=0, x2=0, y2=0):
    length = np.sqrt( (x2-x1)**2 + (y2-y1)**2 )
    return length
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solution_mnedal(x):
    y = []
    
    for xelement in x:
        
        if xelement < -1*np.pi:
            y.append(np.sin(xelement) - 1)
        
        elif xelement >= -1*np.pi and xelement <= np.pi:
            y.append(xelement/np.pi)
        
        elif xelement > np.pi:
            y.append(-1*np.cos(xelement))
        
        else:
            print('Invalid entry!')
    
    plt.figure()
    plt.plot(x, y, '.-')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Solution')
    plt.xlim(-8, 8)
    plt.ylim(-1, 1)
    plt.grid()
    plt.show()
    
    return y
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = solution_mnedal(x)
```

```python
# Specify the columns you want to filter
columns_to_filter = [
    'Disturbance Y/M/D (UT) (a)',
    'ICME Plasma/Field Start, End Y/M/D (UT) (b)',
    'dV (km/s) (h)',
    'V_ICME (km/s) (i)',
    'V_max (km/s) (j)',
    'B (nT) (k)',
    'Dst (nT) (m)',
    'V_transit (km/s) (n)',
    'LASCO CME Y/M/D (UT) (o)']

# Check if the columns are present in the DataFrame
filtered_columns = [col for col in columns_to_filter if col in df.columns]

# Filter the DataFrame using the filtered columns
filtered_df = df[filtered_columns]

# Reset index
filtered_df = filtered_df.reset_index()
filtered_df.drop(columns='index', axis=1, inplace=True)
```

```python
filtered_df
```

```python
# Rename columns
filtered_df = filtered_df.rename(columns={
    'Disturbance Y/M/D (UT) (a)': 'Disturbance Y/M/D (UT)',
    'ICME Plasma/Field Start, End Y/M/D (UT) (b)': 'ICME Plasma/Field Start (UT)',
    'ICME Plasma/Field Start, End Y/M/D (UT) (b)': 'ICME Plasma/Field End (UT)',
    'dV (km/s) (h)': 'dV (km/s)',
    'V_ICME (km/s) (i)': 'V_ICME (km/s)',
    'V_max (km/s) (j)': 'V_max (km/s)',
    'B (nT) (k)': 'B (nT)',
    'Dst (nT) (m)': 'Dst (nT)',
    'V_transit (km/s) (n)': 'V_transit (km/s)',
    'LASCO CME Y/M/D (UT) (o)': 'LASCO CME Y/M/D (UT)'
})
```

```python
filtered_df
```

```python
# drop rows with those values at that column
filtered_df.dropna(subset=['LASCO CME Y/M/D (UT)'], inplace=True)
filtered_df = filtered_df[filtered_df['LASCO CME Y/M/D (UT)'] != '...']
filtered_df = filtered_df[filtered_df['LASCO CME Y/M/D (UT)'] != '&nbsp']
filtered_df = filtered_df[filtered_df['LASCO CME Y/M/D (UT)'] != 'dg']
```

```python
filtered_df
```

```python
filtered_df['Dst (nT)'] = filtered_df['Dst (nT)'].str.replace('[Q]', '', regex=True)
filtered_df['Dst (nT)'] = filtered_df['Dst (nT)'].str.replace('[P]', '', regex=True)
```
